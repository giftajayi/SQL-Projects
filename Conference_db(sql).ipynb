The project involves the creation and management of a database system for a conference using SQLite, along with a Relational Entity Diagram (ERD) to visualize the database structure and relationships. The database includes tables for storing information about attendees, registrations, feedback, speakers, and sessions.

Conference ER diagram.png

```python
# Loading the SQL extension to create the database.

import sqlite3
%load_ext sql
%sql sqlite:///conference.db
```

```python
# Creating the tables.

%%sql

CREATE TABLE Attendees(
  attendee_id INT PRIMARY KEY NOT NULL,
  attendee_name TEXT NOT NULL,
  attendee_email TEXT NOT NULL,
  attendee_type TEXT NOT NULL
);
```

```python
%%sql
PRAGMA table_info(Attendees);
```

```python
%%sql
CREATE TABLE Registrations(
  reg_id INT PRIMARY KEY NOT NULL,
  reg_date DATE NOT NULL,
  reg_payment_status TEXT NOT NULL,
  attendee_id INT NOT NULL,
  session_id INT NOT NULL,
  FOREIGN KEY (attendee_id) REFERENCES Attendees(attendee_id),
  FOREIGN KEY (session_id) REFERENCES Sessions(session_id)
);




```

```python
%%sql
PRAGMA table_info(Registrations);
```

```python
%%sql
CREATE TABLE Feedback(
  feedback_id INT PRIMARY KEY NOT NULL,
  feedback_title TEXT NOT NULL,
  session_description TEXT NOT NULL,
  session_date DATE NOT NULL,
  session_time TIME NOT NULL
);
```

```python
%%sql
PRAGMA table_info(Feedback);
```

```python
%%sql
CREATE TABLE Speakers(
  speaker_id INT PRIMARY KEY NOT NULL,
  speaker_name TEXT NOT NULL,
  speaker_bio TEXT NOT NULL,
  session_id INT NOT NULL,
  FOREIGN KEY (session_id) REFERENCES Sessions(session_id)
);
```

```python
%%sql
PRAGMA table_info(Speakers);
```

```python
%%sql
CREATE TABLE Sessions(
  session_id INT PRIMARY KEY NOT NULL,
  session_title TEXT NOT NULL,
  session_description TEXT NOT NULL,
  session_date DATE NOT NULL,
  session_time TIME NOT NULL
);
```

```python
%%sql
PRAGMA table_info(Sessions);
```

Retrieving all attendees

```python
%%sql

SELECT * FROM attendees;
```

Retrieving all sessions scheduled for June 1st.

```python
%%sql

SELECT * FROM sessions WHERE session_date = '2024-06-01';
```

Retrieving the names of attendees who are professionals

```python
%%sql

SELECT attendee_name FROM Attendees WHERE attendee_type = 'Professional';

```

List all sessions along with their speakers

```python
%%sql

SELECT s.session_title, sp.speaker_name
FROM Sessions s
JOIN Speakers sp ON s.session_id = sp.session_id;

```

Finding the most popular session based on registrations:

```python
%%sql

SELECT s.session_title, COUNT(r.reg_id) AS registration_count
FROM Sessions s
JOIN Registrations r ON s.session_id = r.session_id
GROUP BY s.session_title
ORDER BY registration_count DESC
LIMIT 1;
```

Retrieve sessions along with the total number of feedback entries.

```python
%%sql

SELECT s.session_title, COUNT(f.feedback_id) AS total_feedback
FROM Sessions s
LEFT JOIN Feedback f ON s.session_date = f.session_date AND s.session_time = f.session_time
GROUP BY s.session_title;
```

Find the sessions attended by a specific attendee, Alice Johnson.

```python
%%sql

SELECT s.session_title
FROM Sessions s
JOIN Registrations r ON s.session_id = r.session_id
JOIN Attendees a ON r.attendee_id = a.attendee_id
WHERE a.attendee_name = 'Alice Johnson';
```

Find speakers who have given sessions on multiple topics.

```python
%%sql

SELECT sp.speaker_name, COUNT(DISTINCT s.session_title) AS topic_count
FROM Speakers sp
JOIN Sessions s ON sp.session_id = s.session_id
GROUP BY sp.speaker_name
HAVING COUNT(DISTINCT s.session_title) > 1;
```

Finding the average number of attendees per session

```python
%%sql

SELECT AVG(session_attendees) AS average_attendees
FROM (SELECT session_id, COUNT(attendee_id) AS session_attendees
      FROM Registrations
      GROUP BY session_id) AS session_counts;
```

Finding the sessions that have no registrations

```python
%%sql

SELECT Sessions.session_title
FROM Sessions
LEFT JOIN Registrations ON Sessions.session_id = Registrations.session_id
WHERE Registrations.session_id IS NULL;
```
